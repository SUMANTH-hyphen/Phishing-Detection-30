## Feature Selection Technique1(SelectKBest with mutual information)

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif
import pandas as pd
# Load the dataset
data = pd.read_csv("phishing.csv")

# Remove 'Index' column
data = data.drop('Index', axis=1)

# Define features and target variable
X = data.drop('class', axis=1)  # Features
y = data['class']  # Target variable

# Define the number of features you want to select
k = 15  # Change this number based on your preference

# Feature selection using mutual information
selector = SelectKBest(mutual_info_classif, k=k)
X_new = selector.fit_transform(X, y)

# Get the indices of the selected features
selected_features_indices = selector.get_support(indices=True)

# Get the corresponding column names of the selected features
selected_column_names = X.columns[selected_features_indices]

# Print the column names of the selected features
for i in selected_column_names:
    print(i)
```

## Feature Selection Technique2(chi-squared test )

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
data = pd.read_csv("phishing.csv")

# Remove 'Index' column if it exists
if 'Index' in data.columns:
    data = data.drop('Index', axis=1)

# Define features and target variable
X = data.drop('class', axis=1)  # Features
y = data['class']  # Target variable

# Ensure non-negative values by scaling to [0, 1] range
scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# Feature selection using chi-squared test
selector = SelectKBest(chi2)
X_new = selector.fit_transform(X_scaled, y)

# Get the scores and p-values of the features
scores = selector.scores_
p_values = selector.pvalues_

# Create a DataFrame to store scores and p-values for each feature
features_info = pd.DataFrame({'Feature': X.columns, 'Score': scores, 'P-value': p_values})

# Sort features by their scores in descending order
features_info = features_info.sort_values(by='Score', ascending=False)

# Print the features ranked by their scores
print("Feature scores and p-values:")
print(features_info)
```
